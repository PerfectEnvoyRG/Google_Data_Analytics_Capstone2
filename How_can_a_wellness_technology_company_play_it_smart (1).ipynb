Therefore, we plan to do the following data manipulation:

```python
# Converting ActivityDate from object to datatime64 dtype
# Converting format of ActivityDate to yyyy-mm-dd.
#df["ActivtiyDate"] = df["ActivityDate"].dt.strftime('%Y-%m-%d')
df["ActivityDate"] = pd.to_datetime(df["ActivityDate"], format="%m/%d/%Y")

# Then, printing head to confirm whether it has been updated to datatime64 dtype and dates to yyyy-mm-dd.
df.info()
df.head(10)
```

```python
#df = df.drop('ActivityDate', axis=1)
df.head(10)
```

```python
# we re-arrange the orders of columns
rearranged_cols = ['Id', "ActivityDate", "DayOfTheWeek", "TotalSteps", "TotalDistance", "TrackerDistance", "LoggedActivitiesDistance", "VeryActiveDistance", "ModeratelyActiveDistance", "LightActiveDistance", "SedentaryActiveDistance", "VeryActiveMinutes", "FairlyActiveMinutes", "LightlyActiveMinutes", "SedentaryMinutes", "TotalMins", "TotalHours", "Calories"]
df = df.reindex(columns=rearranged_cols)
df.head(10)
```

```python
# Create new column Day_Of_The_Week by separating the date into day of the week for further analysis.
df["DayOfTheWeek"] = df["ActivityDate"].dt.day_name()
df.head(10)
```

```python
# Convert all column names to lowercase.
df.columns = df.columns.str.lower()
df.head()
```

```python
# To create a new column named "totalmins" in a DataFrame, containing the sum of 
# "VeryActiveMinutes", "FairlyActiveMinutes", "LightlyActiveMinutes", and "SedentaryMinutes"
df['totalmins'] = df[['veryactiveminutes', 'fairlyactiveminutes', 'lightlyactiveminutes', 'sedentaryminutes']].sum(axis=1)
df.head(10)
```

```python
# Create new column named TotalHours by converting new column "TotalMins" to number of hours.
df['totalhours'] = df['totalmins'] / 60.
df.head()
```

```python
df['totalhours'] = df['totalhours'].astype(float).round(2)
df.head()
```

```python
df.info()
```

```python
# replace every unique the values in cols 'id' by nature number
nature_ids = df['id'].unique()
replacement_dict = {value: idx+1 for idx, value in enumerate(nature_ids)}
df['id'] = df['id'].replace(replacement_dict)
df.head(10)
```

```python
# import matplotlib package
import matplotlib.pyplot as plt

# use historgram to analyze the number of times users logged in app across the week
plt.figure(figsize=(8, 6))
plt.hist(df['dayoftheweek'], bins=7, width = 0.6,edgecolor='black')
plt.xlabel('Day of the Week')
plt.ylabel('Frequency')
plt.title('Login Counts across the Week')
plt.xticks(range(7), ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
plt.show()
```

The frequency of usage throughout the week

```python
# Group data by 'dayoftheweek' and calculate the sum of 'sedentaryminutes' for each day
sedentary_minutes_agg = df.groupby('dayoftheweek')['sedentaryminutes'].sum()
# Set the order of the days of the week from Monday to Sunday
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
# Reorder the grouped data based on the specified day order
sedentary_minutes_agg = sedentary_minutes_agg.reindex(day_order)
print(sedentary_minutes_agg)

# Create a line chart
plt.figure(figsize=(10, 6))
plt.plot(sedentary_minutes_agg.index, sedentary_minutes_agg.values, marker='o')
plt.title('Trend in Aggregated Sedentary Minutes by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Aggregated Sedentary Minutes')
plt.grid(True)
plt.show()
```

```python

# plotting scatter plot
# Set the default style
plt.style.use("default")

# specify size of the chart
plt.figure(figsize=(8,6)) 

plt.scatter(df.totalsteps, df.calories, 
            alpha = 0.8, c = df.calories, 
            cmap = "Spectral")

# add annotations and visuals
median_calories = 2303
median_steps = 7637

plt.colorbar(orientation = "vertical")
plt.axvline(median_steps, color = "Blue", label = "Median steps")
plt.axhline(median_calories, color = "Red", label = "Median calories burned")
plt.xlabel("Steps taken")
plt.ylabel("Calories burned")
plt.title("Calories burned for every step taken")
plt.grid(True)
plt.legend()
plt.show()
```

**Based on the above "calories burned" vesus "every step taken" scatter plot, we can understand:**
1. The more steps you take, the more calories you burn. This means that there is a positive correlation between the number of steps and the calories expended.
2. After taking 15,000 steps, the calorie burning efficiency decreases. This indicates the presence of diminishing marginal returns, where the effect gradually diminishes.
3. We have also noticed some outliers. One data point shows that a user walked over 35,000 steps but consumed less than 3,000 calories, while another data point shows that a user walked less than 20,000 steps but excessively burned nearly 5,000 calories. We infer that these discrepancies may arise from issues in the measurement system of the electronic device.

```python
plt.style.use("default")
plt.figure(figsize=(8,6)) 
# Specify size of the chart
plt.scatter(df.totalhours, df.calories, 
            alpha = 0.8, c = df.calories, 
            cmap = "Spectral")

# adding annotations and visuals
median_calories = 2303
median_hours = 20
median_sedentary = 991 / 60

plt.colorbar(orientation = "vertical")
plt.axvline(median_hours, color = "Blue", label = "Median steps")
plt.axvline(median_sedentary, color = "Purple", label = "Median sedentary")
plt.axhline(median_calories, color = "Red", label = "Median hours")
plt.xlabel("Hours logged")
plt.ylabel("Calories burned")
plt.title("Calories burned for every hour logged")
plt.legend()
plt.grid(True)
plt.show()
```

The above scatter plot is illustrating:
1. There is only a very weak positive correlation between the duration of user login to the app or wearing electronic devices to track activity duration and the calories burned.
2. From the purple vertical line, we can also see that the main reason for this weak positive correlation is that sedentary hours account for the majority of the time.
3. There is one extremely exaggerated outlier in the data, where a user logged into the app for around 25 hours but hardly burned any calories. This is highly unlikely to occur. We can infer that the same reason as mentioned earlier applies here, indicating a possible malfunction in the recording feature of the device.
